```python
#Read source data and print a few records to know a bit of the content.

import pandas as pd
df = pd.read_csv('Absenteeism_at_work.csv', sep=',')
df.head()
```

# 7.1 - Renombrando columnas
El objetivo es convertir las columnas en minúsculas y reemplazar los espacios en blanco o caracteres extraños en las etiquetas de las columnas con guiones bajos.

```python
#Obtains the column names from the dataframe and prints the list

aux_colnames = list(df.columns)
aux_colnames
```

```python
#Using list comprehension turns every column name to lower case, 
#   and replaces all white spaces (' ') with underscore ('_').
#   Important: It does not take in count any special character existing in column names.

aux_colnames = list([ col_name.replace(' ', '_').lower() for col_name in aux_colnames ])
aux_colnames
```

```python
#Finally, rename all column names in the dataframe using the list with renamed columns,
#   and print first 5 records to verify column names are now lower case and no white spaces.

df.columns = aux_colnames
df.head()
```
